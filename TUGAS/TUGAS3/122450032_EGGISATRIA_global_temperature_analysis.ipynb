# Analisis dan Visualisasi Tren Perubahan Suhu Global

Dataset: Global Land-Ocean Temperature Index (1880 - 2023) (https://raw.githubusercontent.com/EgiStr/VisualDataInformation_collage/refs/heads/main/TUGAS/TUGAS3/land_ocean_temperature_index.csv)

SUMBER:(https://data.giss.nasa.gov/gistemp/graphs/graph_data/Global_Mean_Estimates_based_on_Land_and_Ocean_Data/graph.txt)

**Tujuan:** Menunjukkan tren perubahan suhu global dari waktu ke waktu untuk meningkatkan kesadaran akan perubahan iklim.

**User:** Publik umum, mahasiswa lingkungan, atau peneliti.

**Fungsi:** Memberikan informasi yang mudah dipahami tentang tren kenaikan suhu global dengan menggunakan grafik sederhana, seperti grafik garis (line chart) atau heatmap.

**Tone:** Serius dan informatif, menekankan urgensi masalah perubahan iklim.


```python
# Import pustaka yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = 'https://raw.githubusercontent.com/EgiStr/VisualDataInformation_collage/refs/heads/main/TUGAS/TUGAS3/land_ocean_temperature_index.csv'
data = pd.read_csv(file_path)

# Tampilkan 5 baris pertama untuk memahami struktur data
data.head()
```

### Grafik Garis Perubahan Suhu Global

Grafik garis ini menunjukkan tren perubahan suhu global dari tahun 1880 hingga 2023. Terdapat dua jenis data:
- **Tanpa Pemulusan**: Data mentah anomali suhu global.
- **Lowess (5 tahun)**: Data yang telah dihaluskan dengan pemulusan Lowess untuk menunjukkan tren jangka panjang.

Grafik ini bertujuan untuk memperlihatkan kenaikan suhu dari waktu ke waktu dan memberikan gambaran yang lebih jelas tentang bagaimana perubahan iklim berdampak pada suhu global. Pemulusan data membantu kita memahami pola jangka panjang dengan lebih baik.


```python
# Set style untuk grafik
sns.set(style="whitegrid")

# Buat grafik garis
plt.figure(figsize=(12, 7))

# Plot Tanpa Pemulusan
plt.plot(data['Year'], data['No_Smoothing'], label='Tanpa Pemulusan', color='blue', linestyle='-', linewidth=2)

# Plot Lowess(5)
plt.plot(data['Year'], data['Lowess(5)'], label='Lowess (5 tahun)', color='red', linestyle='--', linewidth=2)

# Tambahkan label dan judul
plt.title('Indeks Suhu Daratan-Lautan Global (1880 - 2023)', fontsize=16)
plt.xlabel('Tahun', fontsize=14)
plt.ylabel('Anomali Suhu (C)', fontsize=14)

# Tambahkan grid
plt.grid(True)

# Tambahkan legenda
plt.legend(loc='upper left', fontsize=12)

# Tampilkan plot
plt.tight_layout()
plt.show()
```

### Grafik Batang Perubahan Suhu Tahunan

Grafik ini menunjukkan perubahan tahunan dalam anomali suhu global. Ini adalah visualisasi perubahan dari tahun ke tahun, sehingga kita dapat melihat periode di mana terjadi peningkatan atau penurunan besar dalam suhu global.

Tujuan dari grafik ini adalah untuk memberikan pemahaman yang lebih dalam tentang seberapa cepat suhu berubah di tiap tahunnya.


```python
# Hitung perubahan tahunan
data['Annual_Change'] = data['No_Smoothing'].diff()

# Buat grafik batang untuk perubahan tahunan
plt.figure(figsize=(12, 7))

# Plot perubahan tahunan
plt.bar(data['Year'], data['Annual_Change'], color='green', width=0.8)

# Tambahkan label dan judul
plt.title('Perubahan Tahunan Anomali Suhu Daratan-Lautan Global (1880 - 2023)', fontsize=16)
plt.xlabel('Tahun', fontsize=14)
plt.ylabel('Perubahan Tahunan (C)', fontsize=14)

# Tambahkan garis horizontal di 0
plt.axhline(0, color='black', linestyle='--', linewidth=1)

# Tampilkan plot
plt.tight_layout()
plt.show()
```
